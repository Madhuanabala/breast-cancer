```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyRegressor
```

```python
X = df.drop('bioactivity_class', axis=1)
Y = df.bioactivity_class
```

```python
X.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
X = selection.fit_transform(X)
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
clf = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)
models_train,predictions_train = clf.fit(X_train, X_train, Y_train, Y_train)
models_test,predictions_test = clf.fit(X_train, X_test, Y_train, Y_test)
```

```python
from sklearn.metrics import (
    confusion_matrix, accuracy_score, precision_score, recall_score,
    f1_score, balanced_accuracy_score, matthews_corrcoef, roc_curve, auc,
    mean_absolute_error, cohen_kappa_score
)
from lazypredict.Supervised import LazyClassifier

# LazyClassifier setup
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_train, predictions_train = clf.fit(X_train, X_train, Y_train, Y_train)
models_test, predictions_test = clf.fit(X_train, X_test, Y_train, Y_test)

# Dictionary to store validation metrics
validation_metrics = {
    "Model": [],
    "Accuracy": [],
    "Sensitivity (Recall)": [],
    "Specificity": [],
    "F1 Score": [],
    "Balanced Accuracy": [],
    "MCC": [],
    "ROC AUC": [],
    "MAE": [],
    "Kappa": []
}

# Loop through all models in LazyPredict
for model_name, model in clf.models.items():
    # Train the model and make predictions
    clf_model = model  # Model object
    clf_model.fit(X_train, Y_train)
    y_pred = clf_model.predict(X_test)
    y_prob = clf_model.predict_proba(X_test)[:, 1] if hasattr(clf_model, "predict_proba") else y_pred
    # Convert Y_test and y_pred to integers
    Y_test = Y_test.astype(int)
    try:
        y_prob = y_prob.astype(float)
    except ValueError:
        y_prob = y_prob.astype(int)

    y_pred = y_prob.astype(int)

    # Confusion matrix components
    cm = confusion_matrix(Y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    # Add metrics for this model
    validation_metrics["Model"].append(model_name)
    validation_metrics["Accuracy"].append(accuracy_score(Y_test, y_pred))
    validation_metrics["Sensitivity (Recall)"].append(recall_score(Y_test, y_pred))
    validation_metrics["Specificity"].append(specificity)
    validation_metrics["F1 Score"].append(f1_score(Y_test, y_pred))
    validation_metrics["Balanced Accuracy"].append(balanced_accuracy_score(Y_test, y_pred))
    validation_metrics["MCC"].append(matthews_corrcoef(Y_test, y_pred))

    # ROC and AUC
    fpr, tpr, _ = roc_curve(Y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    validation_metrics["ROC AUC"].append(roc_auc)

    # MAE and Kappa
    validation_metrics["MAE"].append(mean_absolute_error(Y_test, y_pred))
    validation_metrics["Kappa"].append(cohen_kappa_score(Y_test, y_pred))

# Convert metrics to DataFrame
metrics_df = pd.DataFrame(validation_metrics)

# Display metrics
print(metrics_df)

# Plot ROC curves for all models
plt.figure(figsize=(10, 8))
for model_name, model in clf.models.items():
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_pred
    fpr, tpr, _ = roc_curve(Y_test, y_prob)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc(fpr, tpr):.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for All Models')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, mean_absolute_error

# Dictionary to store results
positive_negative_counts = {"Model": [], "Positive": [], "Negative": []}

for model_name, model in clf.models.items():
    # Train the model and make predictions
    clf_model = model
    clf_model.fit(X_train, Y_train)
    y_pred = clf_model.predict(X_test)
    y_prob = clf_model.predict_proba(X_test)[:, 1] if hasattr(clf_model, "predict_proba") else y_pred

    # Convert predictions to integers if necessary
    y_pred = list(map(int, y_pred))

    # Count positive and negative predictions
    positive_count = sum(y_pred)
    negative_count = len(y_pred) - positive_count

    # Store counts
    positive_negative_counts["Model"].append(model_name)
    positive_negative_counts["Positive"].append(positive_count)
    positive_negative_counts["Negative"].append(negative_count)



# Convert counts to DataFrame
counts_df = pd.DataFrame(positive_negative_counts)

# Display the counts for all models
print(counts_df)

# Plot positive and negative counts for all models
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")
ax = sns.barplot(data=counts_df.melt(id_vars="Model"), x="Model", y="value", hue="variable")
ax.set_title("Positive and Negative Prediction Counts for Each Model")
ax.set_xlabel("Model")
ax.set_ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Prediction Type")
plt.show()
```
