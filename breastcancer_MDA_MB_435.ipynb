```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('MDA-MB-435')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df=pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.standard_type.unique()
```

```python
df.to_csv('breast cancer bioactivity_data.csv', index=False)
```

```python
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna()
```

```python
print("\nNumber of Duplicate Rows:", df_cleaned.duplicated().sum())
```

```python
df = df[df.standard_value.notna()]
df
```

```python
bioactivity_class = []
for i in df.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
mol_cid = []
for i in df.molecule_chembl_id:
  mol_cid.append(i)
```

```python
canonical_smiles = []
for i in df.canonical_smiles:
  canonical_smiles.append(i)
```

```python
standard_value = []
for i in df.standard_value:
  standard_value.append(i)
```

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
df
```

```python
df.to_csv(' breast cancer bioactivity_preprocessed_data.csv', index=False)
```
